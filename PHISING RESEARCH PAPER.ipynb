```python
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 

from scipy.io import arff
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix

from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import Perceptron
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
```

```python
ph= arff.loadarff('Training Dataset.arff')
phis = pd.DataFrame(ph[0])
phis.head()
```

```python
for i in phis:
    temp = []
    for j in phis[i]:
        temp.append(int(j))
    phis[i] = temp
phis.head()
```

* SPLITTING DATA *

```python
Y = phis['Result']
X = phis.drop('Result', axis=1) 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

CART

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

PERCEPTRON

```python
print("Accuracy:" , accuracy_score(y_test, y_pred))
```

```python
pcp = Perceptron(max_iter = 1000, eta0=0.005, random_state=0)
```

```python
pcp.fit(X_train, y_train)
```

```python
y_pred = pcp.predict(X_test)
print(round(accuracy_score(y_test, y_pred.round())*100,2))
```

RANDOM FOREST

```python
rndm_frst = RandomForestClassifier(n_estimators = 900 , criterion ='gini', random_state = 0)
```

```python
rndm_frst.fit(X_train, y_train)
```

```python
y_pred = rndm_frst.predict(X_test)
print(round(accuracy_score(y_test,y_pred)*100,2))
```

LOGISTIC REGRESSION

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

Comparing Different Splits

```python
result = pd.DataFrame({
    'Model' : ['Logistic Regression','Random Forest','Perceptron', 'CART']
})
```

```python
for i in range(5, 0, -1):
    col = 'Split-' + str((10-i)*10) +':' + str(i*10)
    lst = []
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=i/10)
    lr.fit(X_train, y_train)
    y_pred = lr.predict(X_test)
    lst.append(round(accuracy_score(y_test, y_pred.round())*100, 2))
    rndm_frst.fit(X_train, y_train)
    y_pred = rndm_frst.predict(X_test)
    lst.append(round(accuracy_score(y_test, y_pred)*100, 2))
    pcp.fit(X_train, y_train)
    y_pred = pcp.predict(X_test)
    lst.append(round(accuracy_score(y_test, y_pred.round())*100, 2))
    classifier.fit(X_train, y_train) 
    y_pred = classifier.predict(X_test)
    lst.append(round(accuracy_score(y_test, y_pred)*100, 2))
    result[col] = lst
```
